```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("population_final.csv")
print(df.shape)
df.head()
df.info()
```

```python
top25 = df.sort_values(by="2022 Population", ascending=False).head(25)

plt.figure(figsize=(12,6))
plt.bar(top25["Country"], top25["2022 Population"])

plt.xticks(rotation=90)

plt.title("Top 25 Countries by Population")
plt.xlabel("Country")
plt.ylabel("2022 Population")

plt.show()
```

```python
top25 = df.sort_values(by="2022 Population", ascending=False).head(25)

# Create figure
fig, ax1 = plt.subplots(figsize=(16,8))

# Bar graph for population
ax1.bar(top25["Country"], top25["2022 Population"], color="lightblue")
ax1.set_xlabel("Country")
ax1.set_ylabel("Population")

# Make x labels vertical
ax1.tick_params(axis='x', rotation=90)

# Second axis for Median Age
ax2 = ax1.twinx()
ax2.plot(top25["Country"], top25["Med Age"], color="red", marker="o")
ax2.set_ylabel("Median Age")

# Title
plt.title("Population and Median Age of Top 25 Countries")

plt.tight_layout()

plt.show()
```

```python
low_age = df.sort_values(by="Med Age").head(10)

plt.figure(figsize=(12,6))

x = low_age["Country"]

plt.bar(x, low_age["2022 Population"], label="Population")
plt.bar(x, low_age["Urban Pop"], label="Urban Population")
plt.bar(x, low_age["Land Area (km2)"], label="Land Area")

plt.title("Population, Urban Population and Land Area")
plt.xlabel("Country")
plt.ylabel("Values")

plt.xticks(rotation=45)

plt.legend()

plt.show()
```

```python
sorted_age = df.sort_values(by="Med Age")

plt.figure(figsize=(12,6))

plt.plot(sorted_age["Country"], sorted_age["Med Age"], 'o')

plt.title("Median Age of Countries (Increasing Order)")
plt.xlabel("Country")
plt.ylabel("Median Age")

plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(12,8))

# 1
plt.subplot(2,2,1)
plt.scatter(df["2022 Population"], df["Land Area (km2)"])
plt.xlabel("Population")
plt.ylabel("Land Area")
plt.title("Population vs Land Area")

# 2
plt.subplot(2,2,2)
plt.scatter(df["Density (per km2)"], df["Land Area (km2)"])
plt.xlabel("Density")
plt.ylabel("Land Area")
plt.title("Density vs Land Area")

# 3
plt.subplot(2,2,3)
plt.scatter(df["Urban Pop"], df["World Share (%)"])
plt.xlabel("Urban Population")
plt.ylabel("World Share")
plt.title("Urban Pop vs World Share")

# 4
plt.subplot(2,2,4)
plt.scatter(df["2022 Population"], df["Urban Pop"])
plt.xlabel("Population")
plt.ylabel("Urban Population")
plt.title("Population vs Urban Pop")

plt.tight_layout()

plt.show()
```

## Exp-2 Regenerate this graph using the matplotlib library

```python

labels = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9', 'A10']
values = [25.5, 33.5, 42.5, 44, 34.5, 62, 28.5, 34.5, 29, 20]
x_pos = [1, 2, 3, 5, 6, 7, 9, 10, 11, 12]

patterns = ['||', '\\\\', '//', '++', '--', '..', '***', 'xx', 'oo', 'OO']
colors = ['red', 'green', 'blue', 'red', 'green', 'blue', 'red', 'green', 'blue', 'red']

fig, ax = plt.subplots(figsize=(10, 5))

for i in range(len(labels)):
    ax.bar(x_pos[i], values[i], 
           color='white',          
           edgecolor=colors[i],      
           hatch=patterns[i],
           width=0.8, 
           linewidth=1.2)

ax.set_xlim(0, 13)
ax.set_ylim(0, 65)

ax.tick_params(top=True, right=True, left=True, bottom=True, direction='in', length=6)

ax.set_xticks(x_pos)
ax.set_xticklabels(labels, rotation='vertical', fontsize=12)
ax.set_yticks([0, 10, 20, 30, 40, 50, 60])
ax.set_yticklabels([0, 10, 20, 30, 40, 50, 60], fontsize=12)

for spine in ax.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1) 

plt.tight_layout()
plt.show()
```
