## Importing libraries 

```python
import pandas as pd
import numpy as np
```

```python
columns = ["symboling","normalized-losses","make","fuel-type","aspiration",
           "num-of-doors","body-style","drive-wheels","engine-location",
           "wheel-base","length","width","height","curb-weight",
           "engine-type","num-of-cylinders","engine-size","fuel-system",
           "bore","stroke","compression-ratio","horsepower","peak-rpm",
           "city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv("imports-85.data",names=columns)
df.replace("?", np.nan, inplace=True)
```

### 1- Total Records Count

```python
print("Total Records:", df.shape[0])
```

### 2-Number of Attributes & Naming List

```python
print("Total Attributes:", df.shape[1])
print("Attribute Names:", df.columns.tolist())
```

### 3- Display All Invalid Records

```python
df[df.isnull().any(axis=1)]
```

### 4- Remove All Invalid Records

```python
df_clean = df.dropna().copy()
print(df_clean)
```

### 5- Average Height, Width, Length

```python
df_clean.loc["height"] = df_clean["height"].astype(float)
df_clean.loc["width"] = df_clean["width"].astype(float)
df_clean.loc["length"] = df_clean["length"].astype(float)
```

```python
print("Average Height:", df_clean["height"].mean())
print("Average Width:", df_clean["width"].mean())
print("Average Length:", df_clean["length"].mean())
```

### 6- Standard Deviation Height, Width, Length

```python
print("Std Height:", df_clean["height"].std())
print("Std Width:", df_clean["width"].std())
print("Std Length:", df_clean["length"].std())
```

### 7- Describe Each Attribute

```python
# describe()- automatically min, max, mean, std, count show karta hai
print(df_clean.describe(include='all'))
```

### 8- Sort Vehicles by Body Style

```python
sorted_df = df_clean.sort_values(by="body-style")
print(sorted_df[["make","body-style"]])
```

### 9- Create Dataset “sunset-1”

```python
sunset_1 = df_clean[["make","fuel-type","num-of-doors",
                     "drive-wheels","engine-location"]]

print(sunset_1.head())
```

```python
sunset_2 = df_clean[["wheel-base","length","width",
                     "height","curb-weight"]]

print(sunset_2.head())
```
