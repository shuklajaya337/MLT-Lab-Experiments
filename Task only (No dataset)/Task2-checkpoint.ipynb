## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1- Load the dataset and create the new data frame as MyDF

```python
columns = ["symboling","normalized-losses","make","fuel-type","aspiration",
           "num-of-doors","body-style","drive-wheels","engine-location",
           "wheel-base","length","width","height","curb-weight",
           "engine-type","num-of-cylinders","engine-size","fuel-system",
           "bore","stroke","compression-ratio","horsepower","peak-rpm",
           "city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv("imports-85.data", names=columns)
df.replace("?", np.nan, inplace=True)
```

```python
MyDF = df[["symboling","fuel-type","body-style","drive-wheels",
                 "wheel-base","length","width","height",
                 "num-of-cylinders","bore","compression-ratio",
                 "horsepower","city-mpg","highway-mpg","price"]].copy()
MyDF.head()
```

### 2- Remove all the invalid samples

```python
MyDF_clean = MyDF.replace("?", pd.NA)
MyDF_clean = MyDF.dropna()
print("Total Records After Cleaning:", MyDF.shape[0])
```

```python
MyDF_clean.isnull().sum()
```

### 3- Visualize the distribution of all the attributes. Plot Histogram for each attribute

```python
num_cols = ["wheel-base","length","width","height",
            "bore","compression-ratio","horsepower",
            "city-mpg","highway-mpg","price"]

for col in num_cols:
    MyDF[col] = MyDF[col].astype(float)
```

```python
MyDF[num_cols].hist(figsize=(15,10))
plt.show()
```

### 4- Distribution Analysis 

#### Bar Graph

```python
cat_cols = ["symboling","fuel-type","body-style",
            "drive-wheels","num-of-cylinders"]

for col in cat_cols:
    MyDF[col].value_counts().plot(kind="bar")
    plt.title(col)
    plt.show()
```

#### Box plot

```python
for col in num_cols:
    MyDF.boxplot(column=col)
    plt.title(col)
    plt.show()
```

### 5- Visualize average height, width and length for each body style

```python
MyDF.groupby("body-style")[["height","width","length"]].mean().plot(kind="bar")
plt.title("Average Size per Body Style")
plt.show()
```

### 6-  Visualize average height, width and length for each fuel-type

```python
MyDF.groupby("fuel-type")[["height","width","length"]].mean().plot(kind="bar")
plt.title("Average Size per Fuel Type")
plt.show()
```

### 7-  Visualize maximum height, width and length for num-of-cylinders

```python
MyDF.groupby("num-of-cylinders")[["height","width","length"]].max().plot(kind="bar")
plt.title("Maximum Size per Cylinders")
plt.show()
```

### 8- Visualize the vehicle distributions in terms of their varying prices

```python
plt.hist(MyDF["price"], bins=10)
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```
