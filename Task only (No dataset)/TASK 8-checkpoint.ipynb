```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(6,4))
plt.xlim(0,10)
plt.ylim(1,5)
plt.text(1.5,4,"Computer Science & Engineering, NIT Patna", fontsize=12)
plt.text(3.5,3,r"$x^2 + y^2 = a^2$", fontsize=16)
plt.text(3.6,2,r"$S=\sum_{i=1}^{n} x^i$", fontsize=16)

plt.show()
```

```python
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)

plt.figure(figsize=(6,4))

plt.plot(t, s, color='blue')

plt.ylim(-2, 2)
plt.xlim(0,5)

plt.title("Cosine Wave")

# annotation
plt.annotate('local max',
             xy=(2.5, -1),        # point where arrow points
             xytext=(3.2, -1.6),  # location of text
             arrowprops=dict(facecolor='green', shrink=0.05))

plt.show()
```

```python
x = [1,3,5,7,9]
y = [2,4,6,8,10]

labels = ["Apple","Mango","Banana","Pineapple","Pear"]

plt.figure(figsize=(4,4))

plt.scatter(x,y,color="blue")

for i, label in enumerate(labels):
    plt.annotate(label,
                 xy=(x[i], y[i]),
                 xytext=(x[i]+1.5, y[i]),
                 arrowprops=dict(arrowstyle="->"))

plt.xlim(0,15)
plt.ylim(1,11)

plt.show()
```

```python
data = [[66, 60, 67, 60, 68],
        [80, 93, 96, 92, 88]]

AAA = data[0]
CCC = data[1]

x = [1,2,3,4,5]

plt.figure(figsize=(7,4))

# Plot AAA
plt.plot(x, AAA, color='red', marker='s', linewidth=2, label='AAA')

# Plot CCC
plt.plot(x, CCC, color='blue', marker='^', linestyle='--', linewidth=2, label='CCC')

# Annotation showing difference
plt.annotate('15%',
             xy=(4,60),
             xytext=(4,93),
             arrowprops=dict(arrowstyle='<->', color='gray'),
             ha='center')

plt.legend()

plt.ylim(50,105)
plt.xlim(0.8,5.2)

plt.show()
```

```python
from matplotlib.patches import Ellipse

data = [[76,65,77,70,78],[80,83,86,82,88]]

companyA = data[0]
companyB = data[1]

x = np.arange(1,6)

plt.figure(figsize=(9,5))

plt.plot(x, companyA,
         color='green',
         marker='s',
         markersize=10,
         linewidth=3,
         label='Company-A')

plt.plot(x, companyB,
         color='blue',
         marker='^',
         markersize=10,
         linestyle='--',
         linewidth=3,
         label='Company-B')

plt.grid(True, linewidth=1, alpha=0.6)

# Ellipse highlight
ellipse = Ellipse((2,75),
                  width=0.5,
                  height=30,
                  color='red',
                  alpha=0.25)

plt.gca().add_patch(ellipse)

plt.text(1.75,90,'18% profit\ndifference',fontsize=12)

plt.xlabel("Year")
plt.ylabel("Profit (%)")

plt.xlim(0.8,5.2)
plt.ylim(50,105)

plt.xticks(np.arange(1,5.1,0.5))

plt.legend(loc='lower center', ncol=2)

plt.show()
```
