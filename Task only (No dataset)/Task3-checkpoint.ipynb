## Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 1- Load Dataset & Create MyDF

```python
df = pd.read_csv("imports-85.data", header=None)
MyDF = df[[0,3,6,7,9,10,11,12,15,18,20,21,23,24,25]].copy()
```

```python
MyDF.columns = ["symboling","fuel-type","body-style","drive-wheels",
                "wheel-base","length","width","height",
                "num-of-cylinders","bore","compression-ratio",
                "horsepower","city-mpg","highway-mpg","price"]
print(MyDF.head())
```

### 2-Remove all the invalid samples

```python
MyDF_clean = MyDF.replace("?", pd.NA)
MyDF_clean = MyDF.dropna()
print("Total Records After Cleaning:", MyDF.shape[0])
```

```python
MyDF.isnull().sum()
```

### 3- Visualize the vehicles' height, width, and length in a single graph

```python
plt.figure(figsize=(9,5))
plt.plot(MyDF["height"])
plt.plot(MyDF["width"])
plt.plot(MyDF["length"])

plt.title("Height, Width, Length")
plt.legend(["Height","Width","Length"])
plt.show()
```

### 4- Visualize the vehicles' height, width, and length in a single graph

```python
plt.figure(figsize=(10,6))
plt.plot(MyDF["height"], linestyle='-', marker='o')
plt.plot(MyDF["width"], linestyle='--', marker='s')
plt.plot(MyDF["length"], linestyle=':', marker='^')

plt.title("Different Line & Marker Style")
plt.legend(["Height","Width","Length"])
plt.show()
```

### 5-Visualize the vehicles' height, width, and length in the increasing order of their price

```python
plt.figure(figsize=(10,6))
MyDF_sorted = MyDF.sort_values("price")

plt.plot(MyDF_sorted["height"])
plt.plot(MyDF_sorted["width"])
plt.plot(MyDF_sorted["length"])

plt.title("Size in Increasing Order of Price")
plt.legend(["Height","Width","Length"])
plt.show()
```

### 6- Visualize the relationship between height and width using a scatter plot

```python
plt.figure(figsize=(10,6))
plt.scatter(MyDF["height"], MyDF["width"])
plt.title("Height vs Width")
plt.xlabel("Height")
plt.ylabel("Width")
plt.show()
```
