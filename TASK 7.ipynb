## Task 1 — Generate the Given Graph

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
import matplotlib.pyplot as plt

# Labels and values
labels = ['A1','A2','A3','A4','A5','A6','A7','A8','A9','A10']
values = [25,34,42,44,35,62,28,35,29,20]

# Custom positions to create gaps between groups
x = [0,1,2, 4,5,6, 8,9,10,11]

# Colors and hatch patterns
colors = ['red','green','blue','red','green','blue','red','green','blue','red']
patterns = ['|','\\','/','+','-','.','*','x','o','O']

plt.figure(figsize=(10,6))

bars = plt.bar(x, values, color='lightgrey', edgecolor=colors, linewidth=2)

# Apply hatch patterns
for bar, pattern in zip(bars, patterns):
    bar.set_hatch(pattern)

# X-axis labels
plt.xticks(x, labels, rotation=90)

# Grid similar to image
plt.grid(True, linestyle='--', alpha=0.6)

plt.ylabel("Value")
plt.title("Patterned Bar Graph")

plt.tight_layout()
plt.show()
```
