```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("daily-min-temperatures.csv")

data['Date'] = pd.to_datetime(data['Date'])

data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Quarter'] = data['Date'].dt.quarter

print(data.head(15))
```

### 1- Visualize the day-wise data from 1981 to 1990

```python
plt.figure(figsize=(14,6))

plt.plot(data['Date'], data['Temp'], color='blue')

plt.title("Daily Minimum Temperature (1981–1990)")
plt.xlabel("Date")
plt.ylabel("Temperature")

plt.show()
```

### 2-  Visualize the average month-wise data from 1981 to 1990

```python
data = pd.read_csv("daily-min-temperatures.csv")

data['Date'] = pd.to_datetime(data['Date'])

data['Month'] = data['Date'].dt.month

monthly_avg = data.groupby('Month')['Temp'].mean()

plt.figure(figsize=(10,5))

plt.plot(monthly_avg.index, monthly_avg.values,
         marker='o', color='green')

plt.title("Average Month-wise Temperature (1981-1990)")
plt.xlabel("Month")
plt.ylabel("Temperature")

plt.xticks(range(1,13))

plt.grid(True)

plt.show()

# The correct approach for this wording is usually groupby, because it shows average temperature for each month.
```

### 3- Visualize the average quarter-wise data from 1981 to 1990 and highlight the highest temperature peaks of the quarter with annotation

```python
data = pd.read_csv("daily-min-temperatures.csv")

# Convert Date column to datetime
data['Date'] = pd.to_datetime(data['Date'])

# Extract Quarter
data['Quarter'] = data['Date'].dt.quarter

# Calculate average temperature quarter-wise
quarter_avg = data.groupby('Quarter')['Temp'].mean()

plt.figure(figsize=(8,5))

plt.plot(quarter_avg.index, quarter_avg.values,
         marker='o', color='blue')

# Find highest temperature peak
peak_temp = quarter_avg.max()
peak_quarter = quarter_avg.idxmax()

# Annotate the peak
plt.annotate(f'Peak: {peak_temp:.2f}',
             xy=(peak_quarter, peak_temp),
             xytext=(peak_quarter, peak_temp+1),
             arrowprops=dict(facecolor='black'))

plt.title("Average Quarter-wise Temperature (1981–1990)")
plt.xlabel("Quarter")
plt.ylabel("Average Temperature")

plt.xticks([1,2,3,4])

plt.grid(True)

plt.show()
```

### 4-Visualize the minimum daily temperature distribution through violin plot year-wise from 1981 to 1990

```python
data = pd.read_csv("daily-min-temperatures.csv")

data['Date'] = pd.to_datetime(data['Date'])
data['Year'] = data['Date'].dt.year

plt.figure(figsize=(12,6))

sns.violinplot(x='Year', y='Temp', data=data)

plt.title("Year-wise Minimum Daily Temperature Distribution (1981–1990)")
plt.xlabel("Year")
plt.ylabel("Temperature")

plt.show()
```

```python
# Load image (change path if needed)
path = r"C:\Users\shukl\OneDrive\College work\.ipynb_checkpoints\.ipynb_checkpoints\book-composition-with-open-book_23-2147690555.avif"
img = Image.open(path)

# Convert image to numpy array
img = np.array(img)

# Separate channels
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]

# Create 3x3 subplot
fig, ax = plt.subplots(3,3, figsize=(10,10))

# -------- First Row : RGB Channels --------
ax[0,0].imshow(R, cmap='Reds')
ax[0,0].set_title("Red Channel")
ax[0,0].axis('off')

ax[0,1].imshow(G, cmap='Greens')
ax[0,1].set_title("Green Channel")
ax[0,1].axis('off')

ax[0,2].imshow(B, cmap='Blues')
ax[0,2].set_title("Blue Channel")
ax[0,2].axis('off')

# -------- Second Row : Histograms --------
ax[1,0].hist(R.ravel(), bins=256, color='red')
ax[1,0].set_title("Red Histogram")

ax[1,1].hist(G.ravel(), bins=256, color='green')
ax[1,1].set_title("Green Histogram")

ax[1,2].hist(B.ravel(), bins=256, color='blue')
ax[1,2].set_title("Blue Histogram")

# Hide third row (not required)
for i in range(3):
    ax[2,i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from PIL import Image
from matplotlib.patches import Ellipse

# Load India map
path = r"C:\Users\shukl\OneDrive\College work\.ipynb_checkpoints\.ipynb_checkpoints\.ipynb_checkpoints\India.gif"
img = Image.open(path)

# Get image size
width, height = img.size

plt.figure(figsize=(6,8))
plt.imshow(img)

# Approximate location of Uttar Pradesh
x = width * 0.40
y = height * 0.32

# Highlight region using ellipse
ellipse = Ellipse((x, y), width*0.15, height*0.08,
                  edgecolor='red',
                  facecolor='none',
                  linewidth=3)

plt.gca().add_patch(ellipse)

# Label
plt.text(x+10, y, "Uttar Pradesh", color="red", fontsize=12)

plt.title("Map of India Highlighting Uttar Pradesh Region")
plt.axis('off')

plt.show()
```
